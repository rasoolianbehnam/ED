```python
%load_ext autoreload
%load_ext Cython
%autoreload 1
import pandas as pd
import numpy as np
import tensorflow as tf
import sys, os
import requests
import time
```

```python
cache = os.path.join(os.environ['HOME'], '.deeplearning4j/uci_synthetic_control')
```

```python
if not os.path.exists(cache):
    os.mkdir(cache)
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/synthetic_control-mld/synthetic_control.data"
    fileName = url.split("/")[-1]
    absPath = os.path.join(cache, fileName)
    print("Downloading file...")
    res = requests.get(url)
    res.raise_for_status()
    f = open(absPath, 'wb')
    chunkSize = 100000
    totalSize = 0
    for chunk in res.iter_content(chunkSize):
        f.write(chunk)
        totalSize += chunkSize
        print("Downloaded %4.2f Mb"%(totalSize/1024/1024), end='\r')
    f.close()
    print("")
    print("Finished downloading.")
```
